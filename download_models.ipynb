```python
model_ids =     [
                "FacebookAI/roberta-base",
                "Xuhui/ToxDect-roberta-large",
                "diptanu/fBERT",
                "GroNLP/hateBERT"
                ]
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import os
```

```python

for model_id in model_ids:
    model_path = "./" + model_id.split("/")[-1]
    print(model_path)
# Download and save tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    tokenizer.save_pretrained(model_path)

    # Download and save the entire model (including pytorch_model.bin)
    model = AutoModelForSequenceClassification.from_pretrained(model_id, use_safetensors=False)
    model.save_pretrained(model_path)
```
